**a)** Faça um gráfico de dispersão (scatter plot) com a variável *YearsAtCompany* no eixo X e a variável *DailyRate* no eixo Y. Coloque o título "YearsAtCompany x DailyRate" e determine o tamanho da imagem como (10,5).

```python
df1.plot(x='YearsAtCompany', y='DailyRate', title='YearsAtCompany x DailyRate', figsize=(10,5), kind='scatter')
```

**b)** Repita o gráfico do item a, identificando as observações do gênero feminino com a cor vermelha e do gênero masculino com a cor verde. Não esqueça de inserir legenda para as cores.

```python
ax1 = df1[df1['Gender'] == 'Female'].plot(x='YearsAtCompany', y='DailyRate', title='YearsAtCompany x DailyRate',
                                          figsize=(10,5), kind='scatter', color='red', label='Feminino')
ax2 = df1[df1['Gender'] == 'Male'].plot(x='YearsAtCompany', y='DailyRate', title='YearsAtCompany x DailyRate',
                                        figsize=(10,5), kind='scatter', color='blue', ax=ax1, label='Masculino')

ax1.set_xlabel('Idade', size = 24)
```

**c)** Repita o gráfico do item a, porém agora identificando os funcionários pela variável 'JobSatisfaction',

```python
ax1 = df1[df1['JobSatisfaction'] == 1].plot(x='YearsAtCompany', y='DailyRate', title='YearsAtCompany x DailyRate',
                                          figsize=(10,5), kind='scatter', color='#FFE4E1', label='Low')
ax2 = df1[df1['JobSatisfaction'] == 2].plot(x='YearsAtCompany', y='DailyRate', title='YearsAtCompany x DailyRate',
                                        figsize=(10,5), kind='scatter', color='#FFB6C1', ax=ax1, label='Medium')
ax1 = df1[df1['JobSatisfaction'] == 3].plot(x='YearsAtCompany', y='DailyRate', title='YearsAtCompany x DailyRate',
                                          figsize=(10,5), kind='scatter', color='#FF69B4', label='High', ax=ax1)
ax2 = df1[df1['JobSatisfaction'] == 4].plot(x='YearsAtCompany', y='DailyRate', title='YearsAtCompany x DailyRate',
                                        figsize=(10,5), kind='scatter', color='#C71585', ax=ax1, label='Very High')
```
